# Part I. ETL Pipeline for Pre-Processing the Files

## PLEASE RUN THE FOLLOWING CODE FOR PRE-PROCESSING THE FILES

#### Import Python packages 

```python
# Import Python packages 
import pandas as pd
import cassandra
import re
import os
import glob
import numpy as np
import json
import csv
```

#### Creating list of filepaths to process original event csv data files

```python
# checking your current working directory
print(os.getcwd())

# Get your current folder and subfolder event data
filepath = os.getcwd() + '/event_data'

# Create a for loop to create a list of files and collect each filepath
for root, dirs, files in os.walk(filepath):
    
# join the file path and roots with the subdirectories using glob
    file_path_list = glob.glob(os.path.join(root,'*'))
```

#### Processing the files to create the data file csv that will be used for Apache Casssandra tables

```python
# initiating an empty list of rows that will be generated from each file
full_data_rows_list = [] 
    
# for every filepath in the file path list 
for f in file_path_list:

# reading csv file 
    with open(f, 'r', encoding = 'utf8', newline='') as csvfile: 
        # creating a csv reader object 
        csvreader = csv.reader(csvfile) 
        next(csvreader)
        
 # extracting each data row one by one and append it        
        for line in csvreader:
            full_data_rows_list.append(line) 

# creating a smaller event data csv file called event_datafile_full csv that will be used to insert data into the \
# Apache Cassandra tables
csv.register_dialect('myDialect', quoting=csv.QUOTE_ALL, skipinitialspace=True)

with open('event_datafile_new.csv', 'w', encoding = 'utf8', newline='') as f:
    writer = csv.writer(f, dialect='myDialect')
    writer.writerow(['artist','firstName','gender','itemInSession','lastName','length',\
                'level','location','sessionId','song','userId'])
    for row in full_data_rows_list:
        if (row[0] == ''):
            continue
        writer.writerow((row[0], row[2], row[3], row[4], row[5], row[6], row[7], row[8], row[12], row[13], row[16]))
```

```python
# check the number of rows in your csv file
with open('event_datafile_new.csv', 'r', encoding = 'utf8') as f:
    print(sum(1 for line in f))
```

## Begin writing your Apache Cassandra code in the cells below

#### Creating a Cluster

```python
# This should make a connection to a Cassandra instance your local machine 
# (127.0.0.1)

from cassandra.cluster import Cluster
cluster = Cluster()

# To establish connection and begin executing queries, need a session
session = cluster.connect()
```

#### Create Keyspace

```python
try:
    session.execute("""
    CREATE KEYSPACE IF NOT EXISTS sparkify 
    WITH REPLICATION = 
    { 'class' : 'SimpleStrategy', 'replication_factor' : 1 }"""
)

except Exception as e:
    print(e)
```

#### Set Keyspace

```python
try:
    session.set_keyspace('sparkify')
except Exception as e:
    print(e)
```

### Now we need to create tables to run the following queries. Remember, with Apache Cassandra you model the database tables on the queries you want to run.

## Create queries to ask the following three questions of the data

### 1. Give me the artist, song title and song's length in the music app history that was heard during  sessionId = 338, and itemInSession  = 4


### 2. Give me only the following: name of artist, song (sorted by itemInSession) and user (first and last name) for userid = 10, sessionid = 182
    

### 3. Give me every user name (first and last) in my music app history who listened to the song 'All Hands Against His Own'




```python
## This table is created for query 1. 
## This Table is created to provide query for finding out what song has been listened to for a particular sessionId
## This table is able to support query when user provide sessionId or both sessionId and itemInSession. sessionId is the partition key where item_in_session is the clustering key. Redundant brackets to make it specific and clear.

try:
    session.execute("CREATE TABLE IF NOT EXISTS app_history_session(session_id int, item_in_session int, artist text, song_title text, length decimal, PRIMARY KEY((session_id), item_in_session))"
)
except Exception as e:
    print(e)

                    
```

```python
## Insert into cassandra app_history_session table based on denormalized event_datafile_new file.
file = 'event_datafile_new.csv'

with open(file, encoding = 'utf8') as f:
    csvreader = csv.reader(f)
    next(csvreader) 
    for line in csvreader:
        query = "INSERT INTO app_history_session(session_id, item_in_session, artist, song_title, length) "
        query = query + " VALUES(%s, %s, %s, %s, %s)"
        session.execute(query, (int(line[8]), int(line[3]), line[0], line[9], float(line[5])))
```

#### Do a SELECT to verify that the data have been inserted into each table

```python
## Try select all to see if records has been inserted correctly.
try:
    rows = session.execute("SELECT * FROM app_history_session");
except Exception as e:
    print(e)
    
for row in rows:
    print (row.session_id, row.item_in_session, row.artist, row.song_title, row.length);
```

```python
## Try out the query intended.
try:
    rows = session.execute("SELECT * FROM app_history_session WHERE session_id=797 AND item_in_session=1");
except Exception as e:
    print(e)
    
for row in rows:
    print (row.session_id, row.item_in_session, row.artist, row.song_title, row.length);
```

### COPY AND REPEAT THE ABOVE THREE CELLS FOR EACH OF THE THREE QUESTIONS


```python
## This table is created for query 2. 
## This Table is created to provide query for finding out what the user has been listened to for a particular sessionId
## An example is to provide detailed history playlist for user when they click on a particular session activity on their app.
## This table is able to support query when provided by sessionId AND user_id. The query will not work if only sessionId is provided as user_id is part of the partition_key for this table model. 
## The reason for putting user_id to be part of primary key instead of clustering key is to reduce the partition key cardinality and thus increase query performance, as sessionId grows quite rapidly. In the event we need to query by sessionId only, we can use app_history_session table.

try:
    session.execute("CREATE TABLE IF NOT EXISTS app_history_artist(session_id int,  user_id int, item_in_session int, artist text, song_title text, first_name text, last_name text, PRIMARY KEY((session_id, user_id), item_in_session)) WITH CLUSTERING ORDER BY (item_in_session ASC);"
)
except Exception as e:
    print(e)
```

```python

file = 'event_datafile_new.csv'

with open(file, encoding = 'utf8') as f:
    csvreader = csv.reader(f)
    next(csvreader) # skip header
    for line in csvreader:
        query = "INSERT INTO app_history_artist(session_id, user_id, item_in_session, artist, song_title, first_name, last_name) "
        query = query + " VALUES(%s, %s, %s, %s, %s, %s, %s)"
        session.execute(query, (int(line[8]), int(line[10]), int(line[3]), line[0], line[9], line[1], line[4]))
                  
```

```python
## Try print out to see if inserted correctly.
try:
    rows = session.execute("SELECT * FROM app_history_artist");
except Exception as e:
    print(e)
    
for row in rows:
    print (row.artist, row.song_title, row.first_name, row.last_name);  
```

```python
## Trying out query. 
try:
    rows = session.execute("SELECT * FROM app_history_artist WHERE user_id=26 AND session_id=169");
except Exception as e:
    print(e)
    
for row in rows:
    print (row.artist, row.song_title, row.first_name, row.last_name);   
```

```python
## This table is created for query 3. 
## This Table is created to provide query for finding out who has been listening to a particular song.
## song is the partition key and user_id is the clustering key. Clustering order is added by asc to provide easy analysis of data.
## Additional brackets and ordering keyword is added for clarity, even though they may not be needed as default impl will auto infer user_id as a clustering key.

try:
    session.execute("CREATE TABLE IF NOT EXISTS app_history_user(song text, user_id int, first_name text, last_name text,PRIMARY KEY((song), user_id)) WITH CLUSTERING ORDER BY (user_id ASC);"
)
except Exception as e:
    print(e)
```

```python

file = 'event_datafile_new.csv'

with open(file, encoding = 'utf8') as f:
    csvreader = csv.reader(f)
    next(csvreader) # skip header
    for line in csvreader:
        query = "INSERT INTO app_history_user(song, user_id, first_name, last_name) "
        query = query + " VALUES(%s, %s, %s, %s)"
        session.execute(query, (line[9], int(line[10]), line[1], line[4]))

try:
    rows = session.execute("SELECT * FROM app_history_user");
except Exception as e:
    print(e)
    
for row in rows:
    print ( row.song, row.first_name, row.last_name);                    
                    
```

```python
## Trying out query
try:
    rows = session.execute("SELECT * FROM app_history_user WHERE song='Horn Concerto No. 4 in E flat K495: II. Romance (Andante cantabile)'");
except Exception as e:
    print(e)
    
for row in rows:
    print (  row.song, row.first_name, row.last_name,);  
```

### Drop the tables before closing out the sessions

```python

try:
    session.execute("DROP TABLE app_history_user")
    session.execute("DROP TABLE app_history_artist")
    session.execute("DROP TABLE app_history_session")
except Exception as e:
    print(e)
```

### Close the session and cluster connection¶

```python
session.shutdown()
cluster.shutdown()
```
